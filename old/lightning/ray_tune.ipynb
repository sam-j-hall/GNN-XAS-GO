```python
# Random seed to reproduce data split
generator = torch.Generator().manual_seed(11)
# Split dataset into train, validation and test sets
train_data, val_data, test_data = random_split(dataset, [0.75, 0.125, 0.125], generator)

print(f'Length of training data: {len(train_data)}')
print(f'Length of validation data: {len(val_data)}')
print(f'Length of testing data: {len(test_data)}')

# torch.save(test_data, './processed/test_data_coronene.pt')
```

```python
# Load data splits into DataLoaders
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
val_loader = DataLoader(val_data, batch_size=40, shuffle=True)

print(f'Number of batches in train loader: {len(train_loader)}')
print(f'Number of batches in val loader: {len(val_loader)}')
```

```python
from ray import tune
# --- Define parameters for GNN model
config = {
    "num_tasks": 200,
    "num_layer": 3,
    "in_channels": [dataset.num_features, 64, 128],
    "out_channels": [64, 128, 256],
    "gnn_type": "gcn",
    "heads": 1,
    "drop_ratio": 0.3,
    "graph_pooling": "mean",
    "lr": tune.loguniform(1e-4, 0.1)
}

model = GNN_model(config)
model
```

```python
import math
import os
def train_mnist_tune(config, num_epochs=10, num_gpus=0):
    model = GNN_model(config)
    trainer = pl.Trainer(
        max_epochs=num_epochs,
        # If fractional GPUs passed in, convert to int.
        gpus=math.ceil(num_gpus),
        logger=TensorBoardLogger(save_dir=os.getcwd(), name="", version="."),
        enable_progress_bar=False,
        callbacks=[
            TuneReportCallback(
                {"loss": "ptl/val_loss", "mean_accuracy": "ptl/val_accuracy"},
                on="validation_end",
            )
        ],
    )
    trainer.fit(model)
```

```python
# --- Seed for reproducability
pl.seed_everything(1994, workers=True)
torch.backends.cudnn.deterministic = True
# torch.backends.cudnn.benchmark = False
# torch.use_deterministic_algorithms=True
torch.set_float32_matmul_precision('high')
```

```python
from pytorch_lightning.callbacks import ModelCheckpoint

checkpoint_callback = ModelCheckpoint(
    filename='epoch={epoch:02d}-val_loss={val_loss:.6f}',
    monitor='val_loss',
    mode='min',
    save_top_k=10
)
```

```python
# --- Path to save model
path = './ml_model'
logger = pl.loggers.TensorBoardLogger(save_dir=path)
# --- Lightining trainer
trainer = pl.Trainer(logger=logger,
                    callbacks=[checkpoint_callback],
                    default_root_dir=path,
                    max_epochs=300, 
                    accelerator='auto',
                    log_every_n_steps=1)
```
